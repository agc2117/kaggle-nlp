```python
clean_reviews = []

for x in train['review']:
    clean_reviews.append(review_to_words(x))
```

```python
len(clean_reviews)
```

```python
#Using Scikit-Learn to create a bag of words for feature modeling
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(analyzer='word',
                            tokenizer = None,
                            preprocessor = None,
                            stop_words = None,
                            max_features = 5000) #Caps the word vector at a length of 5,000, arbitrary
```

```python
train_data_features = vectorizer.fit_transform(clean_reviews)
```

```python
train_data_features = train_data_features.toarray()
```

```python
train_data_features[0].sum()
```

```python
vocab = vectorizer.get_feature_names()
```

```python
print(vocab)
```

```python
import numpy as np
```

```python
dist = np.sum(train_data_features, axis = 0)
```

```python
for tag, count in zip(vocab, dist):
    print(tag, count)
```

```python
#Initialize a Random Forest Model, and train 100 trees on the sentiments
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators = 100)

forest = forest.fit(train_data_features, train['sentiment'])
```
